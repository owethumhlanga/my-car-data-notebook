```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
```

```python
df = pd.read_csv("Car Sample DATA.csv")
```

```python
df.head()
```

```python
df.dtypes
df.describe()
```

```python
# Calculate interquartile range (IQR)
Q1 = df['sellingprice'].quantile(0.25)
Q3 = df['sellingprice'].quantile(0.75)
IQR = Q3 - Q1
print('Interquartile Range (IQR):', IQR)
```

```python
# Calculate coefficient of variation (CV)
mean_val = df['sellingprice'].mean()
std_dev = df['sellingprice'].std()
CV = (std_dev / mean_val) * 100
print('Coefficient of Variation (CV):', CV)
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Visualizations
plt.figure(figsize=(12, 6))
```

```python
#Histogram of a numerical column
plt.subplot(1, 2, 1)
sns.histplot(df['sellingprice'], kde=True)
plt.title('Histogram of Selling Price')
```

```python
#Boxplot of another numerical column
plt.subplot(1, 2, 2)
sns.boxplot(y=df['sellingprice'])
plt.title('Boxplot of Selling Price')

plt.tight_layout()
plt.show()
```

```python
# Density plots
plt.subplot(3, 2, 5)
sns.kdeplot(df['sellingprice'], shade=True, color='skyblue')
plt.title('Density Plot of sellingprice')

plt.subplot(3, 2, 6)
sns.kdeplot(df['year'], shade=True, color='salmon')
plt.title('Density Plot of Year')

plt.tight_layout()
plt.show()
```

```python
# Scatter plot to analyze relationship between two numerical columns
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sellingprice', y='year', data=df, alpha=0.7)
plt.title('Scatter Plot of Selling Price vs Year')
plt.xlabel('sellingprice')
plt.ylabel('year')
plt.show()
```

```python
# Fill missing values in a numeric column with mean
df['sellingprice'] = df['sellingprice'].fillna(df['sellingprice'].mean())
```
